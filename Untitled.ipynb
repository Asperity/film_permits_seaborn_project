```python
import pandas as pd
import seaborn as sns

raw_data = pd.read_csv("Film_Permits_20260713.csv")
```

# The data set that we are using is Film Permits

# The question that we aim to answer is if we can track when certain areas will be have parking restrictions due to filming

# This will help New Yorkers plan ahead of these restrictions

```python
#as there was an outlier in the data, we decided to filter it out from further consideration
borough_data = raw_data[raw_data['Borough'] != "Theater Load in and Load Outs"]
```

```python
#First, we wanted to see how many film restrictions there are per borough
sns.countplot(borough_data, x="Borough")
```

```python
#Next, we removed the rows that did not have a value for their start times or did not hold parking
borough_data = borough_data.dropna(subset=['StartDateTime', 'ParkingHeld'])
```

```python
#Next, we wanted only the times from the datetime columns

borough_data['Start DateTime'] = pd.to_datetime(borough_data['StartDateTime'], unit='s')
borough_data['Only Start Time'] = borough_data['Start DateTime'].dt.time 
```

```python
#This groups the times into resonable distinctions so the graph is more visually discernable 

def time_period(t):
    hour = t.hour

    if 6 <= hour < 12:
        return "Morning (6AM - 11:59PM)"
    elif 12 <= hour < 18:
        return "Afternoon (12PM - 5:59PM)"
    else:
        return "Night (6PM - 5:59AM)"

borough_data["Time Period"] = borough_data["Only Start Time"].apply(time_period)
```

```python
#Finally, we created a chart that displays when these parking restrictions tend to start based on different boroughs
sns.histplot(
    borough_data,
    x='Borough', hue='Time Period',
    multiple='stack'
)
```
